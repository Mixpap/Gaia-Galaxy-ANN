```python
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray
```

```python
def DifIm(Ims,index,gf=0.0):
    P=[rgb2gray(i) for i in Ims]
    P=[gaussian_filter(i,gf) for i in P]
    P=[i.astype(int) for i in P]
    DI1=abs(P[index]-P[index-1])
    DI2=abs(P[index-1]-P[index-2])
    return np.bitwise_and(DI1,DI2)
```
