```python
def make_box(xx,yy):
    """
    Make Boxplot (xarray, yarray)
    """
    db=[]
    for uni in np.unique(xx):
        db.append(yy[np.where(xx==uni)])
    plt.boxplot(db) 
    plt.xticks(range(1,len(np.unique(xx))+1),np.unique(xx))
    return db

def NHU(Hlayers):
    """
    Count Number of Units in ALL Hidden Layers
    """
    s=np.zeros(len(Hlayers))
    i=0
    for j in Hlayers:
        try:
            s[i]=int(j)
        except:
            for k in j:
                s[i]=s[i]+int(k)
        i= i+1
    return s
```

```python
def Folder_Classification_Report(Folder,TP=0.5,FP=0.5):
    i=0
    n=[]
    d=[]
    Hlayers=[]
    TPE=[]
    TPS=[]
    TPI=[]
    TPQ=[]
    accuracy=[]
    filelist=[]
    #Folder='/home/strangeattractor/Projects/ipython/asteroskopeio/Results'
    print 'Reading %d files in %s'%(len(listdir(Folder)),Folder)
    for f in listdir(Folder):
        filelist.append(f)
        datafile='Results/'+f
        ID,real,res=funcs.ReadData(datafile)
        CM,NCM,dfM,dfN = funcs.ConfusionMatrix(ID,real,res,TP,FP)
        #MM,DM,dfMM,dfDM = funcs.AverageMatrix(real,res)

        d.append(float(f[f.find('d')+1:f.find('c')]))
        n.append(float(f[f.find('n')+1:f.find('d')]))
        Hlayers.append([f[j+1:j+3] for j in xrange(len(f)) if f.find('h', j) == j])
        TPE.append(NCM[0,0])
        TPS.append(NCM[1,1])
        TPI.append(NCM[2,2])
        TPQ.append(NCM[3,3])
        accuracy.append((TPE[i]+TPS[i]+TPI[i]+TPQ[i])/NCM.sum())
        print i,filelist[i],accuracy[i]
        i=i+1

    accuracy=np.array(accuracy)
    d=np.array(d)
    n=np.array(n)
    NL = np.array([len(j) for j in Hlayers])
    TPE=np.array(TPE)
    TPS=np.array(TPS)
    TPI=np.array(TPI)
    TPQ=np.array(TPQ)
    DF= pd.DataFrame({'File':filelist,'d':d,'n':n,'Accuracy':accuracy,'Layers':Hlayers,'TPE':TPE,'TPS':TPS,'TPI':TPI,'TPQ':TPQ})
    return DF

def Folder_Regression_Report(Folder):
        
    AVE_A=0.42128425
    STD_A=0.28714299
    AVE_Z=0.09973424
    STD_Z=0.05802179
    
    i=0
    n=[]
    d=[]
    Hlayers=[]
    sdA=[]
    sdZ=[]
    filelist=[]
    #Folder='/home/strangeattractor/Projects/ipython/asteroskopeio/Results'
    files = listdir(Folder)

    del files[files.index('i96h24h12n0.02d0.1c1000o2R11.res')]
    del files[files.index('SnnsOptR13C0TestAX_UgcLib12_L2N1S0G150I0Z0020A0060L.pat')]
    
    print 'Reading %d files in %s'%(len(files),Folder)
    
    for f in files:
        filelist.append(f)
        datafile='AZres/'+f
        ID,real,res=funcs.ReadData(datafile)

        uAn=res[:,0]
        rAn=real[:,0]
        uZn=res[:,1]
        rZn=real[:,1]

        dAn=uAn-rAn
        dZn=uZn-rZn

        uA=3.0*STD_A*uAn+AVE_A
        uZ=3.0*STD_Z*uZn+AVE_Z

        rA=3.0*STD_A*rAn+AVE_A
        rZ=3.0*STD_Z*rZn+AVE_Z

        dA=uA-rA
        dZ=uZ-rZ

        d.append(float(f[f.find('d')+1:f.find('c')]))
        n.append(float(f[f.find('n')+1:f.find('d')]))
        Hlayers.append([f[j+1:j+3] for j in xrange(len(f)) if f.find('h', j) == j])
        sdA.append(dA.std(ddof=1.0))
        sdZ.append(dZ.std(ddof=1.0))
        print i,filelist[i],sdA[i],sdZ[i]
        i=i+1

    d=np.array(d)
    n=np.array(n)
    NL = np.array([len(j) for j in Hlayers])
    sdA=np.array(sdA)
    sdZ=np.array(sdZ)
    DF= pd.DataFrame({'File':filelist,'d':d,'n':n,'SDA':sdA,'SDZ':sdZ,'Layers':Hlayers})
    return DF
```
